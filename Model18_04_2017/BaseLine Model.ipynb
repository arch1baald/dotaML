### Missings

```python
def findColumnsWithNan(d):
    features = list(d)
    containsMissings = []
    for i in range(0, d.shape[1]):
        if d.iloc[:, i].count() < d.shape[0]:
            containsMissings.append(features[i])
    return containsMissings
```

## Data Partition

```python
if target in categorical_columns:
    categorical_columns.remove(target)
    print "Removed from categorical"
if target in numerical_columns:
    numerical_columns.remove(target)
    print "Removed from numerical"
if target in binary_columns:
    binary_columns.remove(target)
    print "Removed from numerical"
```

# Model and Assess

## Decision Tree

```python
from sklearn import tree
from sklearn import metrics
dt = tree.DecisionTreeClassifier(max_depth = 7, min_samples_split = 4)
dt.fit(X_train, y_train)
print "Train accuracy = %s" %metrics.accuracy_score(y_train, dt.predict(X_train))
print "Test accuracy = %s" %metrics.accuracy_score(y_test, dt.predict(X_test))
```

```python
from sklearn import tree
from sklearn.grid_search import GridSearchCV

dt = tree.DecisionTreeClassifier()
clf = GridSearchCV(dt, param_grid={ 'max_depth': [2, 5, 7, 10, 12, 15, 20, 50],
                                   'min_samples_leaf' : [4, 8, 12, 16, 20]
                                  }, 
                   cv = 10)
clf.fit(X_train, y_train)

TreeArccuracy = clf.best_score_
bestTree = clf.best_estimator_

print "The processing is over\n"
print "The Best Cross-Validation accuracy is %s \n" %clf.best_score_
print "The Best max_depth is %s \n" %clf.best_estimator_.max_depth
print "The Best min_samples_leaf is %s \n" %clf.best_estimator_.min_samples_leaf
print "The Best min_samples_split is %s \n" % clf.best_estimator_.min_samples_split
```
