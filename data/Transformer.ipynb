```python
import pandas as pd
import json
```

```python
df_matches = pd.read_csv('analysis_table.csv', header=None, names=['data'])
df_matches['data'] = df_matches.data.apply(lambda json_string: json.loads(json_string))
df_matches['odota'] = df_matches.data.apply(lambda dct: dct['odota']['players'])
# df_matches.drop('data', axis=1, inplace=True) 
df_matches.head()
```

```python
def unpack(lst):
    series_list = []
    for slot in range(10):
        series_list.append(pd.Series(lst[slot]))
    return pd.DataFrame(series_list)
```

```python
matches_list = []
for match in df_matches.odota:
    matches_list.append(unpack(match))
```

```python
df_matches = pd.concat(matches_list)
df_matches.head()
```
